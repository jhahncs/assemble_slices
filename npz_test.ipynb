```python
import numpy as np
import torch
import random 
tt = np.zeros_like(np.arange(190).reshape(1, 190) , bool)
for i in range(190):
    if random.random() > 0.5:
        tt[0][i] = True
    else:
        tt[0][i] = False
tt = torch.tensor(tt )

print(tt.shape,tt)


tt = np.zeros_like(np.arange(190).reshape(1, 190,1) , float)
for i in range(190):
    tt[0][i] =  random.random()
tt = torch.tensor(tt )

print(tt.shape,tt)

```
